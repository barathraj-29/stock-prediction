# DATA VISUALIZATION

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('data.csv')
```

```python
data.head()
```

```python
df=data.dropna()
```

```python
df.columns
```

```python
# Histogram
plt.hist(df['Volume'])
plt.xlabel('Volume')
plt.ylabel('Volume Count')
plt.title('Volume')
```

```python
plt.hist(df['Open'], color='purple')
plt.xlabel('Open')
plt.ylabel('Open Range')
plt.title('Open')
```

```python
sns.countplot(y="company", data=df, palette="husl")
```

```python
def piechart(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(8,8), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 15)

piechart(df, 'company')
```

```python
df.hist(figsize=(15,55),layout=(15,4))
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(25,12))
sns.heatmap(df.corr(),annot=True,fmt='0.2%',cmap='rocket',ax=ax)
#YlGnBu, Reds, tab10, rocket, mako, flare, crest, magma, viridis, cubehelix, YlOrBr, vlag, icefire, Spectral, coolwarm, Blues
```

```python
plt.plot(df['Close'],df['Volume'],color='y')
plt.xlabel('Close')
plt.ylabel('Volume')
plt.title('Closing Value')
plt.show()
```
